```python
df_ratings_dummy = df_ratings.copy().dropna(thresh =20, axis=1).fillna(0)
df_ratings_dummy
```

```python

#cosine similarity of the ratings
similarity_matrix = cosine_similarity(df_ratings_dummy, df_ratings_dummy)
similarity_matrix_df = pd.DataFrame(similarity_matrix, index=df_ratings.index, columns=df_ratings.index)
#similarity_matrix_df.to_excel("similarity_metrix2.xlsx")
similarity_matrix_df
```

```python
#calculate ratings using weighted sum of cosine similarity
#function to calculate ratings
def calculate_ratings(id_course, id_user):
    if id_course  in df_ratings and id_user in similarity_matrix_df:
        cosine_scores = similarity_matrix_df[id_user] #similarity of id_user with every other user
        ratings_scores = df_ratings[id_course] #ratings of every other user for the course id_course
        #won't consider users who havent rated id_course so drop similarity scores and ratings corresponsing to np.nan
        index_not_rated = ratings_scores[ratings_scores.isnull()].index
        ratings_scores = ratings_scores.dropna()
        cosine_scores = cosine_scores.drop(index_not_rated)
        #cosine_scores= cosine_scores.dropna()
        #calculating rating by weighted mean of ratings and cosine scores of the users who have rated the movie
        y = cosine_scores.sum()
       # print (round(y,0))
        if y !=0:
            ratings_movie = np.dot(ratings_scores, cosine_scores)/y
            #ratings_movie = (ratings_scores*cosine_scores).sum()/cosine_scores.sum()
            return ratings_movie*1
        else:
            print("cosine_sum_ contains zero value %.3f  with course id:"  %id_user + id_course)
    else:
        return 2.55
    
```
